```python
from selenium import webdriver
from time import sleep
from bs4 import BeautifulSoup
import os
import re
import pandas as pd
import numpy as np
```

```python
driver = webdriver.Chrome(executable_path="/Users/sufyanahmad/Desktop/DSI 6/materials/classes/week-06/labs/python-webscraping_opentable-lab-master/chromedriver/chromedriver")
```

```python
datajoblinks = []

for x in range(0,230):
   URL = "https://www.mycareersfuture.sg/search?search=Data&sortBy=new_posting_date&page=%d" % x
   driver.get(URL)
   sleep(3)

   for i in range(0,20):
       jobcard_xpath_iter = '//*[@id="job-card-%d"]/div/a' % i
       jobcards = driver.find_elements_by_xpath(jobcard_xpath_iter)

       for jobs in jobcards:
           joblink = jobs.get_attribute('href')
           datajoblinks.append(joblink)
```

```python
# Remove this line to run the code below.
// interrupt
```

```python
company = []
title = []
address = []
jtype = []
experience = []
category = []
salary = []
num_applications = []
post_date = []
expiry_date = []
description = []
requirements = []

for link in datajoblinks:

    driver.get(link)
    sleep(3)

    try:
        company.append(driver.find_element_by_name('company').text)
    except:
        company.append(np.nan)
    
    try:
        title.append(driver.find_element_by_id('job_title').text)
    except:
        title.append(np.nan)
        
    try:
        address.append(driver.find_element_by_id('address').text)
    except:
        address.append(np.nan)
        
    try:
        jtype.append(driver.find_element_by_id('employment_type').text)
    except:
        jtype.append(np.nan)
        
    try:
        experience.append(driver.find_element_by_id('seniority').text)
    except:
        experience.append(np.nan)
        
    try:
        category.append(driver.find_element_by_id('job-categories').text)
    except:
        category.append(np.nan)
        
    try:
        salary.append(driver.find_element_by_class_name('lh-solid').text)
    except:
        salary.append(np.nan)
        
    try:
        num_applications.append(driver.find_element_by_id('num_of_applications').text)
    except:
        num_applications.append(np.nan)
        
    try:
        post_date.append(driver.find_element_by_id('last_posted_date').text)
    except:
        post_date.append(np.nan)
        
    try:
        expiry_date.append(driver.find_element_by_id('expiry_date').text)
    except:
        expiry_date.append(np.nan)
        
    try:
        description.append(driver.find_element_by_id('description-content').text)
    except:
        description.append(np.nan)
        
    try:
        requirements.append(driver.find_element_by_id('requirements-content').text)
    except:
        requirements.append(np.nan)
```

```python
# Remove this line to run the code below.
// interrupt
```

```python
df = pd.DataFrame({'Company': company, "Title": title, "Address": address, "Type": jtype, "Seniority": experience,
                   "Category": category, "Salary": salary, "Num_Applications": num_applications, "Post_Date": post_date, "Expiry_Date": expiry_date,
                   "Description": description, "Requirements": requirements})
```
